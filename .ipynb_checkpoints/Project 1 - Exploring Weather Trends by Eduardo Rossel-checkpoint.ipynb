The main objetive of this work is to analyze trends in global temperature and compare it to the local trends for the temperature in Santiago, Chile.

## Extracting data

The first step is the extraction of data from the database provided. To do so, the following SQL statements where used.

SQL Statements:

`SELECT * FROM global_data;`
* Exploring the city list in order to search for cities in Chile:`SELECT * FROM city_list WHERE country = 'Chile';`
* Extracting the city data from DB for Santiago: `SELECT year, avg_temp FROM city_data WHERE city = 'Santiago' AND country = 'Chile';`

The results of the above queries were download as .csv

## Preparing data for visualization

Pandas, Seaborn and Matplotlib will be used to prepare the data and create the visualizations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline
```

### Importing files and calculating moving averages

The first step in analysis will be to import the .csv files and to add a column to each dataframes with the moving average of the temperature. The moving average temperature is calculated with a 7-year window.

Both global and Santiago's 7-year moving average temperature are shown in Figure 1. Figure 2 shows the same data but with the x label starting on the year 1855 as the data from Santiago starts there. Both figures also show the data of Los Angeles (Chile), Tokyo (Japan) and Paris (France). This data has been added to show is the trends observed in Santiago also take place in other parts of the world. Intuition says it should be this way.

Moving averages where calculated using pandas build in functions. For example: `global_data['7-y MA'] = global_data.iloc[:,1].rolling(window=7).mean()`

```python
# Read CSV Files
global_data = pd.read_csv('global-results.csv')
santiago_data = pd.read_csv('santiago-results.csv')
losangeles_data = pd.read_csv('losangeles-results.csv')
tokyo_data = pd.read_csv('tokyo-results.csv')
paris_data = pd.read_csv('paris-results.csv')
```

```python
# Quick review of the global data
global_data.describe()
```

```python
# Quick review of the Santiago data
santiago_data.describe()
```

```python
# Calculate Moving Average
global_data['7-y MA'] = global_data.iloc[:,1].rolling(window=7).mean()
santiago_data['7-y MA'] = santiago_data.iloc[:,1].rolling(window=7).mean()
losangeles_data['7-y MA'] = losangeles_data.iloc[:,1].rolling(window=7).mean()
tokyo_data['7-y MA'] = tokyo_data.iloc[:,3].rolling(window=7).mean()
paris_data['7-y MA'] = paris_data.iloc[:,3].rolling(window=7).mean()
```

```python
# Plot Moving Average of Global and Santiago Temperature

f = plt.figure(figsize=(15,11))
ax1 = f.add_subplot(211)
ax2 = f.add_subplot(212)

ax1.plot(global_data['year'],global_data['7-y MA'],label='Global')
ax1.plot(santiago_data['year'],santiago_data['7-y MA'],label='Santiago')
ax1.plot(losangeles_data['year'],losangeles_data['7-y MA'],label='Los Angeles, Chile')
ax1.plot(tokyo_data['year'],tokyo_data['7-y MA'],label='Tokyo')
ax1.plot(paris_data['year'],paris_data['7-y MA'],label='Paris')
ax1.set_xlabel('Year')
ax1.set_ylabel('7-year MA Temperature [ºC]')
ax1.set_title('Figure 1. Yearly Average Global Temperature')
ax1.axis(xmin=global_data['year'].min(),xmax=global_data['year'].max())


ax2.plot(global_data['year'],global_data['7-y MA'],label='Global')
ax2.plot(santiago_data['year'],santiago_data['7-y MA'],label='Santiago')
ax2.plot(losangeles_data['year'],losangeles_data['7-y MA'],label='Los Angeles, Chile')
ax2.plot(tokyo_data['year'],tokyo_data['7-y MA'],label='Tokyo')
ax2.plot(paris_data['year'],paris_data['7-y MA'],label='Paris')
ax2.set_xlabel('Year')
ax2.set_ylabel('7-year MA Temperature [ºC]')
ax2.set_title('Figure 2. Yearly Average Global Temperature (Zoom)')
ax2.axis(xmin=santiago_data['year'].min(),xmax=santiago_data['year'].max())

plt.legend(title='City',loc='best',bbox_to_anchor=(1.0, 0.25))
plt.tight_layout()
```

It's possible to calculate the correlation coefficient between the Global and Santiago's temperature. A correlation coefficient of 0.89 is obtained. Given the high correlation between the variables, a linear model could be a good approach into predicting Santiago's temperature given Global Average Temperature value. 

```python
# Calculating correlation coefficients between Global and Santiago MA Temperature
merge_data = global_data[['year','7-y MA']].merge(santiago_data[['year','7-y MA']], on='year', how='left')
mdc = merge_data.corr()
mdc
```

```python
# Get coefficients of linear fit
from scipy import stats
merge_data_dropna = merge_data.dropna()
slope, intercept, r_value, p_value, std_err = stats.linregress(merge_data_dropna['7-y MA_x'],merge_data_dropna['7-y MA_y'])
```

```python
# Plot of linear fit
sns.lmplot(data=merge_data, x='7-y MA_x',y='7-y MA_y',line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope,intercept)})
plt.title('Figure 3. Linear model between Global and Santiago temperature')
plt.xlabel('7-year MA Global temperature [º C]')
plt.ylabel('7-year MA Santiago temperature [º C]')
plt.legend(loc='upper left')
```

# Data analysis

The data from the global temperature shows a tendency to increase over time. Between 1750 and 1850, there are some spikes in the average temperature of the globe. The overall trend is that our planet is getting hotter, since the moving average global temperature has increased in 2 ºC in the last 200 years. This is no surprise as we know that climate change its a phenomenom that has increased since the industrial revolution to modern days. 


The year moving average temperature for Santiago follows a similiar trend to global data. In the last 150 years, the average temperature of the city has increased in 1,5 ºC. From the linear plots we can see that the temperature of Santiago is always below global temperature, the average difference is of 2,8 ºC. Santiago also shows a lower moving average temperature that others cities like Tokyo, Paris and Los Angeles. The latter being rather odd, as Los Angeles is to the south of Santiago so it should have lower average temperatures.

When exploring the correlation between the temperature of Santiago and the Global, it can be noted that there is a high correlation between this two variables. The correlation coefficient between the two is 0,89, thus confirming this statement. Figure 3 also shows confirms that there is a strong correlation. As the average temperature of the planet increases so does the temperature of Chile. This was highly expected.

As there is a high correlation between global temperature and the temperature of Santiago, a linear model could be used to predict the temperature of Santiago. Figure 3. shows the linear model made using available tools for Python.
